```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/all_dataset/Books.csv')
```

```python
df[:7]
```

```python
df.fillna(0,inplace=True)
```

```python
df.drop(['index'], axis=1, inplace=True)
```

```python
df.describe()
```

```python
df[df["Publishing Year"] > 1900]
```

```python
df.dropna(subset='Publishing Year', inplace=True)
```

df.duplicated().sum()

```python
df.nunique()
```

```python
plt.hist(df['Publishing Year'])
plt.xlabel('Publishing Year')
plt.ylabel('Frecuency')
plt.title('Frecuency of Publishing Year')
plt.show()
```

```python
df.groupby('Author')['Book_average_rating'].mean().sort_values(ascending=False)
```

```python
sns.boxplot(x='genre', y='Book_ratings_count', data=df, hue='genre')
plt.ylabel('Book Ratings Count')
plt.xlabel('Genere')
plt.title('Box Plot of Book Rating Count for Each Genere')
plt.show()
```

```python
sns.scatterplot(x='sale price', y='units sold', data=df)
plt.xlabel('Sale Price')
plt.ylabel('Units Sold')
plt.title('Scater Plot of Sale Price vs Units Sold')
plt.show()
```

```python
df.columns
```

```python
language_count = df['language_code'].value_counts()
language_count
```

```python
plt.figure(figsize=(7,7))
plt.pie(language_count, startangle=90)
plt.title('Language Distribution of Books')
plt.legend(labels= language_count.index, title='Languages', loc="center left", bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
df.groupby('Publisher ')['publisher revenue'].sum().sort_values(ascending=True)
```

```python
df.groupby('Author_Rating')['Book_ratings_count'].mean().sort_values(ascending=False)
```

```python
total_gross_sales_author = df.groupby('Author')['gross sales'].sum()
total_gross_sales_author
```

```python
total_gross_sales_author.sort_values(ascending=False).head(10).plot(kind='bar')
plt.ylabel('Total Gross Sales')
plt.xlabel('Author')
plt.title('Total Gross Sales for Each Author')
plt.show()
```

```python
sns.boxplot(x='Author_Rating', y='units sold', data=df, hue='Author_Rating')
plt.ylabel('Units Solds')
plt.xlabel('Author Rating')
plt.title('Box Plot for Units Sold of Each Author Rating')
plt.show()
```

```python
df.groupby('Publishing Year')['units sold'].sum().plot(kind='line')
plt.ylabel('Publishing Year')
plt.xlabel('Units Sold')
plt.title('Total Units Sold Over The Years')
plt.show()
```

```python
df.hist(figsize=(12, 8), bins=30)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Publisher ', y='publisher revenue', data=df, hue='genre', palette='Set2')

plt.title('Publisher Revenue Analysis')
plt.xlabel('Publisher',labelpad=25 )
plt.ylabel('Publisher Revenue (in $)', labelpad=30, verticalalignment='center')

plt.xticks(rotation=65, ha='center')

plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.pairplot(df, hue="Author_Rating", height=3)
plt.show()
```

```python
plt.figure(figsize=(9,7))
sns.heatmap(df[['Publishing Year', 'Book_average_rating', 'Book_ratings_count','gross sales', 'publisher revenue', 'sale price', 'sales rank', 'units sold']].corr(),cmap = 'viridis', annot=True, linewidths=2)
```
